### 16.8 Explanation
In this exercise, I used SQLAlchemy to connect to the SQLite database books.db. I executed a SQL query to select the title column from the book table and sorted the results in alphabetical order using ORDER BY. Finally, I printed each book title.

```python
from sqlalchemy import create_engine, text

engine = create_engine("sqlite:///books.db")

with engine.connect() as conn:
    result = conn.execute(text("SELECT title FROM book ORDER BY title"))
    
    for row in result:
        print(row[0])
```
